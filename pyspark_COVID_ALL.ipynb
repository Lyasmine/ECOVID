```python
#voir le nombre de lignes et colonnes
print ('Shape of training data :', toto1)
```

```python
#Enlever les colonnes qui ne servent à rien 
fr_sante = toto1.drop(["CityCode", "City", "Province", "CountryCode"], axis =1)
fr_sante
```

```python
# Créer une nouvelle fonction qui détermine si la valeur en paramètre est manquante:
def num_missing(x):
    return sum(x.isnull())
# On applique cette fonction pour chaque colonne:
print("Valeurs manquantes par colonne:")
print(fr_sante.apply(num_missing, axis=0)) #axis=0 définit que la fonction sera bien appliquée sur chaque colonne
# Puis application pour chaque ligne:
print("\nValeurs manquantes par ligne:")
print(fr_sante.apply(num_missing, axis=1).head()) #axis=1 définit que la fonction sera bien appliquée sur chaque ligne
```

```python
#Lire le fichier en dataframe

df = pd.DataFrame(fr_sante)
df
```

```python
#nombre de pays 
df["Country"].nunique()

```

```python
#Verification des dates 
df["Date"].unique()
```

```python
#Verification des types de données 
df.dtypes
```

```python
#Verification s'il n y a pas d'erreurs ou des valeurs null
df.isna().sum()
```

```python
#Standariser la date 

from datetime import date
DaTe=df["Date"].apply(lambda x: x.date())
DaTe
```

```python
#Changer la colonne date par la nouvelle 
df["Date"]=DaTe
df
```

```python
df.isna().sum()
```

```python
#Nettoyer le fichier en regroupant par pays et date, le nombre de morts, contaminés, guéris et actifs 

new_df= df.groupby(['Country','Date']).agg({'Confirmed':'sum','Deaths':'sum','Recovered':'sum','Active':'sum'}).reset_index()
new_df
```

```python
#TRansformer le fichier en csv 
new_df.to_csv('out.csv')
```
